```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv('energy_dataset_.csv')
data.head(5)
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data['Type_of_Renewable_Energy'] = data['Type_of_Renewable_Energy'].astype(str)

mapping = {
 '1' : 'Solar',
 '2' : 'Wind',
 '3' : 'Hydroelectric',
 '4' : 'Geothermal',
 '5' : 'Biomass',
 '6' : 'Tidal',
 '7' : 'Wave'
}

data['Type_of_Renewable_Energy'] = data['Type_of_Renewable_Energy'].map(mapping)
data.head(5)
```

```python
data['Funding_Sources'] = data['Funding_Sources'].astype(str)

fund_map = {
 '1' : 'Government',
 '2' : 'Private',
 '3' : 'Public-Private Partnership'
}

data['Funding_Sources'] = data['Funding_Sources'].map(fund_map)
```

```python
data.head(5)
```

```python
sns.set_style('whitegrid') 

plt.figure(figsize=(10, 18)) 

plt.subplot(3, 1, 2)
df_ghg = data.groupby('Type_of_Renewable_Energy')['GHG_Emission_Reduction_tCO2e'].mean().reset_index()
sns.barplot(x='Type_of_Renewable_Energy', y='GHG_Emission_Reduction_tCO2e', data=df_ghg, palette='plasma')
plt.title('Average GHG Emission Reduction (tCO2e) by Type of Renewable Energy')

plt.subplot(3, 1, 3)
df_air = data.groupby('Type_of_Renewable_Energy')['Air_Pollution_Reduction_Index'].mean().reset_index()
sns.barplot(x='Type_of_Renewable_Energy', y='Air_Pollution_Reduction_Index', data=df_air, palette='spring')
plt.title('Average Air Pollution Reduction Index by Type of Renewable Energy') 

plt.xticks(rotation=45) 

plt.tight_layout() 

plt.show()
```

```python
data_ed = data.groupby('Type_of_Renewable_Energy')[['Energy_Production_MWh', 'Energy_Consumption_MWh', 'Energy_Storage_Capacity_MWh']].mean().reset_index()
data_ed
```

```python
ax = data_ed.plot(x='Type_of_Renewable_Energy', kind='bar', stacked=True, figsize=(10,6))

ax.set_title('Average Energy Metrics by Type of Renewable Energy')
ax.set_xlabel('Type of Renewable Energy')
ax.set_ylabel('Average MWh') 
ax.legend(title='Metrics')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Type_of_Renewable_Energy', y='Installed_Capacity_MW', data=data, errorbar=None, palette='spring')
plt.title('Installed Capacity of Different Types of Renewable Energy')
plt.xlabel('Type of Renewable Energy')
plt.ylabel('Installed Capacity (MW)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Type_of_Renewable_Energy', y='Storage_Efficiency_Percentage', data=data, palette='plasma')
plt.title('Storage Efficiency of Different Types of Renewable Energy')
plt.xlabel('Type of Renewable Energy')
plt.ylabel('Storage Efficiency (%)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Type_of_Renewable_Energy', y='Jobs_Created', data=data, errorbar=None, palette='spring')
plt.title('Number of Jobs Created by Different Types of Renewable Energy')
plt.xlabel('Type of Renewable Energy')
plt.ylabel('Number of Jobs Created')
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 2, 1)
sns.barplot(x='Type_of_Renewable_Energy', y='Initial_Investment_USD', data=data, palette='plasma')
plt.title('Initial Investment by Type')
plt.xlabel('Type of Renewable Energy')
plt.ylabel('Initial Investment (USD)')

plt.subplot(1, 2, 2)
sns.barplot(x='Type_of_Renewable_Energy', y='Financial_Incentives_USD', data=data, palette='spring')
plt.title('Financial Incentives by Type')
plt.xlabel('Type of Renewable Energy')
plt.ylabel('Financial Incentives (USD)')

plt.tight_layout()
plt.show()
```
